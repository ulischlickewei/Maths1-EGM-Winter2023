## 1) Limits of functions in Python

```python
from sympy import *
x = symbols('x')
```

In order to compute a limit of a symbolic expression in Python, we use the exact same syntax as for sequences, only that this time we specify another limit point.

```python
f = sin(x)/x
f
```

```python
plot(f,(x,-1,1))
```

```python
limit(f,x,0)
```

```python
g = sin(pi/x)
```

```python
g
```

```python
plot(g,(x,-1,1))
```

```python
limit(g,x,0)
```

We can of course also compute limits which depend on additional parameters:

```python
a = symbols('a')
f2 = a*sin(x)/x
f2
```

```python
limit(f2,x,0)
```

## 2) Derivatives in Python

```python
f = exp(2*sin(x))
f
```

```python
diff(f,x)
```

```python
g = x**(1.5)/3**x
g
```

```python
diff(g,x)
```

Computation of the second derivative:

```python
diff(g,x,x) # differentiate wrt x and then again wrt x --> 2nd derivatnve
```

```python
diff(g,x,2)
```
