#Plotting with SymPy

```python
from sympy import *
x = symbols('x')
f = x**2
plot(f)
```

```python
help(plot)
```

```python
plot(f,(x,-2,2))
```

```python
plot((f,(x,-2,2)),(x**3,(x,-2,2)))
```
