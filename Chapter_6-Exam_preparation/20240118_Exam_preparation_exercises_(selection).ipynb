```python
from sympy import *
x = symbols('x')
f = log(x**2 + 2*x)
fp = diff(f,x)
```

```python
plot(fp, (x,1,2))
```

From the plot we observe that the only integer value which $f'$ takes on $[1,2]$ is $1$. Thus, we have to find a solution to the equation $$f'(x) = 1 \iff f'(x) - 1 = 0$$ on the interval $[1,2]$.

```python
 fp
```

```python
solve(fp-1,x)
```

Thus, the answer is $\sqrt{2}$.
