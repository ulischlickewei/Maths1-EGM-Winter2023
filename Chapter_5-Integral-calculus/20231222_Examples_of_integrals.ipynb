**Step 1:** Compute Taylor polynomial $T_5$ at $x_0=0$ (choose this point of development because this is the only where we can evaluate $\exp$ and its derivatives by hand.

```python
t = symbols('t')
f = exp(-1/2*t**2)
T5 = series(f,t,0,6).removeO()
T5
```

**Step 2:** Compute approximate value of integral

```python
integrate(1/sqrt(2*pi)*T5, (t,0,1)).evalf()
```

```python
# comparison with solution
(5/6+1/40)*1/sqrt(2*pi).evalf()
```

**Step 3:** Estimate the approximation error

*3a)*: Estimate the approximation error of $T_5$ on $[0,1]$

```python
plot(diff(f,t,6),(t,0,1))
```

```python
plot(diff(f,t,7),(t,0,0.2))
```

From the above we conclude that the sixth derivative of our integrand takes its maximal absolute value at $t=0$.

Therefore, Taylor's formula leads to the following estimate for the approximation error on $[0,1]$:
$$M = \left|\frac{f^{(6)}(0)}{6!} (1-0)^6\right|.$$

```python
M = abs(diff(f,t,6).subs(t,0)/factorial(6)*1)
M
```

*3b)*: Estimate the approximation error when integrating

$$\text{err} \leq M \cdot \text{length of interval of integration} = M \cdot (1-0) = M.$$
